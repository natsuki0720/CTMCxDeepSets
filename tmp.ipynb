```python
import pandas as pd
import matplotlib.pyplot as plt

def df_to_image(df, path="df_table.png", dpi=800, font_size=10):
    fig, ax = plt.subplots()
    ax.axis("off")

    tbl = ax.table(
        cellText=df.values,
        colLabels=df.columns,
        cellLoc="center",
        loc="center",
    )
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(font_size)
    tbl.scale(1, 2)  # 行の高さを少し広げる

    plt.tight_layout()
    fig.savefig(path, dpi=dpi, bbox_inches="tight")
    plt.close(fig)


df = pd.read_csv("notebooks/eval_error.csv",)
df.loc[df["name"]=="shoban","name"]="deck"
wide = (
    df.pivot_table(
        index=["name","%","condition_state"],          # 行：同じnameを集約
        columns="metric",  # 列：percentileごとに列化
        values="value",        # 中身：value
        aggfunc="first"        # 重複がある場合の扱い（状況に応じて変更）
    )
    
)

wide.to_csv("error_table.csv",index=True)
import dataframe_image as dfi


# まずこれを試す（環境によって動く/動かないが、playwright回避になることがある）
# dfi.export(wide, "table.png", table_conversion="matplotlib")

# 多少VSCodeっぽく寄せるなら（例）
styled = wide.style.set_table_styles(
    [{"selector":"th, td", "props":[("border","1.2px solid #ddd"), ("padding","7px")]}]
).set_properties(**{"text-align": "center"})

# まずこれを試す（環境によって動く/動かないが、playwright回避になることがある）
dfi.export(wide, "table.png", fontsize=10,table_conversion="matplotlib",dpi=800)
```

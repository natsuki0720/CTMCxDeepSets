```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# plt.rc('font', size=18)  # フォントサイズを 14 に統一

# データを作成
data = {
    "道路橋": [37.0, 54.0, 75.0],
    "トンネル": [25.0, 35.0, 52.0],
    "河川管理施設": [22.0, 42.0, 65.0],
    "水道管路": [9.0, 21.0, 41.0],
    "下水道管渠": [8.0, 16.0, 34.0],
    "港湾施設": [27.0, 44.0, 68.0]
}
df = pd.DataFrame(data)
df = df[["道路橋","河川管理施設", "下水道管渠"]]
# 図の設定
fig, axes = plt.subplots(1, 3, figsize=(6  , 4), sharey=True)

# 各カラムごとに棒グラフを描画
columns = df.columns
bar_width = 0.6
index = np.arange(df.shape[0])
legends = [23, 30,40]
colors = ["#4C566A", "#5E81AC", "#D75F5F"]  # 落ち着きから警告へ移行する3色


for i in range(len(df)):  # 最初の3つのカラムのみ
    axes[i].bar(index, df[columns[i]], width=bar_width,color = colors)
    axes[i].set_title(columns[i],fontsize = 20)
    axes[i].set_xticks(index)
    axes[i].set_xticklabels([legends[i] for i in range(df.shape[0])],fontsize = 12)

fig.text(0.05,0.98,"建設後50年を超えるインフラの割合",va="center",fontsize = 24)
fig.text(0.02, 0.55,"割合（％）",va='center', rotation=90,fontsize = 24)
fig.text(0.5, 0.05, "年度", fontsize=24, va='center')
plt.yticks(np.linspace(0,100,6), fontsize = 10)

# plt.suptitle("建設後50年を超えるインフラの割合")
plt.subplots_adjust(bottom=0.15, top=1, wspace=0.2)
plt.tight_layout(rect=[0.04, 0.06, 0.98, 0.96])  # グラフ配置の微調整
plt.show()
```
